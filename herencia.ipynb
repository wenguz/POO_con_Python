```python
print("PROBLEMA: HERENCIA Y POLIMORFISMO - Wendy Guzman rojas\n")
"""Tomando como base el siguiente diagrama UML, usted deberá 
escribir un programa donde se pondrá en práctica la herencia y 
el polimorfismo.
"""
class Animal:
    """Super clase"""
    
    def setEspecie(self,especie):
        self.especie=especie
        
    def getEspecie(self):
        return self.especie
    
    def printMensaje(self):  
        print("Soy un animal.....")
        
class Perro(Animal):
    """Subclase perro"""
    
    def setRaza(self,r):
        self.raza=r
        
    def getRaza(self):
        print("Soy un perro de raza:",self.raza)
    
    def printMensaje(self):
        # Metodo sobreescrito  
        print("Guau Guau!!")
        
class Pez(Animal):
    """subclase pez"""
    
    def setAlimentacion(self,a):
        self.alimentacion=a
        
    def getAlimentacion(self):
        print("Estoy comiendo:",self.alimentacion)
    
    def printMensaje(self):  
        # Metodo sobreescrito
        print("Gluuuu")
        
class Serpiente(Animal):
    """sublcase serpiente"""
    
    def setVenenosa(self,v):
        self.venenosa=v
        
    def isVenenosa(self):
        if self.venenosa is True:
            print("Soy venenosa")
        else:
            print("No soy venenosa")
    
    def printMensaje(self):  
        # Metodo sobreescrito
        print("ZZZZZZZZZ")
        
        
nuevo_animal1=Perro()
nuevo_animal1.setEspecie("Perro")
print(f"Soy un animal de especie: {nuevo_animal1.getEspecie()}")
nuevo_animal1.printMensaje()
nuevo_animal1.setRaza("Golden")
nuevo_animal1.getRaza()
print("_____________________________")
nuevo_animal2=Pez()
nuevo_animal2.setEspecie("Pez")
print(f"Soy un animal de especie: {nuevo_animal2.getEspecie()}")
nuevo_animal2.printMensaje()
nuevo_animal2.setAlimentacion("copos")
nuevo_animal2.getAlimentacion()
print("_____________________________")
nuevo_animal3=Serpiente()
nuevo_animal3.setEspecie("serpiente")
print(f"Soy un animal de especie: {nuevo_animal3.getEspecie()}")
nuevo_animal3.printMensaje()
nuevo_animal3.setVenenosa(True)
nuevo_animal3.isVenenosa()
```
