```python
print("CREAR 4 CLASES CON SUS VARIABLES Y COMPORTAMIENTOS")
print("Wendy Guzman Rojas")
print("")
print("")

class Animal:
    """Super clase"""
    
    def __init__(self,especie,nombre,genero,edad):
        self.especie=especie
        self.nombre=nombre
        self.genero=genero
        self.edad=edad
        
    def tipo(self,tipo):
        if tipo is True:
            return f"Animal domestico"
        else:
            return f"Animal Salvaje"
        
    def __str__(self):
        return f"Me llamo: {self.nombre} ({self.genero}).\n Soy de la especie {self.especie} y tengo {self.edad} años"
        
class Perro(Animal):
    """Subclase perro"""
    
    def setRaza(self,r):
        self.raza=r
        
    def getRaza(self):
        print("Soy un perro de raza:",self.raza)
    
    def printMensaje(self):
        # Metodo sobreescrito  
        print("Corro")
        
class Pez(Animal):
    """subclase pez"""
    
    def setAlimentacion(self,a):
        self.alimentacion=a
        
    def getAlimentacion(self):
        print("Estoy comiendo:",self.alimentacion)
    
    def printMensaje(self):  
        # Metodo sobreescrito
        print("Nado")
        
class Serpiente(Animal):
    """sublcase serpiente"""
    
    def setVenenosa(self,v):
        self.venenosa=v
        
    def isVenenosa(self):
        if self.venenosa is True:
            print("Soy venenosa")
        else:
            print("No soy venenosa")
    
    def printMensaje(self):  
        # Metodo sobreescrito
        print("Me arrastro")
        
class Ave(Animal):
    """sublcase ave"""
    def volar(self,v):
        if v is True:
            print("Puedo volar")
        else:
            print("No puedo volar")
            
    def printMensaje(self):  
        # Metodo sobreescrito
        print("Salto")
        
print("ANIMALES")
print("")
nuevo_animal1=Perro("Perro","Dayna","Hembra",2)
print(nuevo_animal1)
print(f"{nuevo_animal1.tipo(True)}")
nuevo_animal1.printMensaje()
nuevo_animal1.setRaza("Golden")
nuevo_animal1.getRaza()
print("_____________________________")

nuevo_animal2=Pez("Pez","Dory","Hembra",1)
print(nuevo_animal2)
print(f"{nuevo_animal2.tipo(True)}")
nuevo_animal2.printMensaje()
nuevo_animal2.setAlimentacion("copos")
nuevo_animal2.getAlimentacion()
print("_____________________________")

nuevo_animal3=Serpiente("Serpiente","Naginy","Macho",15)
print(nuevo_animal3)
print(f"{nuevo_animal3.tipo(False)}")
nuevo_animal3.printMensaje()
nuevo_animal3.setVenenosa(True)
nuevo_animal3.isVenenosa()
print("_____________________________")

nuevo_animal4=Ave("Ave","Polly","Hembra",2)
print(nuevo_animal4)
print(f"{nuevo_animal4.tipo(True)}")
nuevo_animal4.printMensaje()
nuevo_animal4.volar(False)
```
